```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

FIRST DATA SOURCE AND CLEANING - stats web scraping

```python
# NBA season we will be analyzing
year = 2019
# URL page we will scraping 
url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html".format(year)

html = urlopen(url)
soup = BeautifulSoup(html)

```

```python

soup.findAll('tr', limit=2)

headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
headers = headers[1:]
headers



rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]

stats = pd.DataFrame(player_stats, columns = headers)
stats
```

```python
stats.columns = stats.columns.str.lower()
stats
```

```python
stats = stats.rename(columns={'fg%':'fgpercent', '3p%':'3ppercent', '2p%':'2ppercent', 'ft%':'ftpercent'})
stats
```

```python
stats.columns
```

```python
del stats["efg%"]
stats.columns
```

SECOND DATA SOURCE AND CLEANING - bb_salaries = TOP_35 - excel

```python
file_to_read = 'Resources/The-Worlds-Highest-Paid-Athletes.xlsx'
high_salaries_df = pd.read_excel(file_to_read)
high_salaries_df.head()
```

```python
high_salaries_basketball = high_salaries_df.loc[high_salaries_df['Sports'] == 'Basketball']
high_salaries_basketball
```

```python
basketball_salaries_df = high_salaries_basketball[['Name', 'Pay', 'Salary/Winnings', 'Endorsement', 'Marital Status']]
basketball_salaries_df = basketball_salaries_df.rename(columns={'Name': 'Player', 'Salary/Winnings': 'Salary_Winnings', 'Marital Status': 'Marital_Status'})
bb_salaries = basketball_salaries_df.rename(columns={'Player': 'player', 'Pay': 'pay', 'Salary_Winnings': 'salary_winnings', 'Endorsement': 'endorsement', 'Marital_Status': 'marital_status'})
bb_salaries
```

THIRD FILE TO CLEAN - bb_salaries_cln = players - CSV

```python
file_to_read = 'Resources/Basketball_Salaries.csv'
bb_salaries_df = pd.read_csv(file_to_read)
bb_salaries_df.head()
```

```python
bb_salaries_cln = bb_salaries_df[['Player', 'Tm', '2019-20']]
bb_salaries_cln
```

```python
bb_salaries_cln['Player'] = bb_salaries_cln['Player'].str.split('\\')
bb_salaries_cln.head()
```

```python
bb_salaries_cln[['Player','Handle']] = pd.DataFrame(bb_salaries_cln.Player.values.tolist(), index= bb_salaries_cln.index)
bb_salaries_cln.head()
```

```python
del bb_salaries_cln['Handle']
bb_salaries_cln 
```

```python
bb_salaries_cln = bb_salaries_cln.rename(columns={'2019-20': 'Current_Salary'})
bb_salaries_cln.columns = bb_salaries_cln.columns.str.lower()
bb_salaries_cln
```

PUTTING FROM PANDAS DFs TO POSTGRES SQL

```python
from sqlalchemy import create_engine
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/basketball_db"
```

```python
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python

bb_salaries.to_sql(name='top_35', con=engine, if_exists='append', index=False)
```

```python
bb_salaries_cln.to_sql(name='players', con=engine, if_exists='append', index=False)
```
